```python
import google.generativeai as genai

api_key = input("Enter Gemini API Key")

genai.configure(api_key=api_key)

model = genai.GenerativeModel("gemini-3.1-flash-lite")

def run_ai_tool(tool_name, user_input):

    if tool_name == "Study Assistant":

        prompt = f"""
        Explain {user_input} in simple language.

        Also provide:
        - Summary
        - Quiz Questions
        - Fun Fact
        """

    elif tool_name == "Notes Summarizer":

        prompt = f"""
        Summarize the following text.

        Provide:
        - Summary
        - Key Points
        - Action Items

        Text:
        {user_input}
        """

    elif tool_name == "Email Writer":

        prompt = f"""
        Write a professional email about:

        {user_input}
        """

    elif tool_name == "Idea Generator":

        prompt = f"""
        Generate 10 creative ideas about:

        {user_input}
        """

    response = model.generate_content(prompt)

    return response.text
```

```python
tool_name = input(
"""
Choose Tool:

Study Assistant
Notes Summarizer
Email Writer
Idea Generator

Your choice:
"""
)

user_input = input(
"Enter your topic or text: "
)

result = run_ai_tool(
    tool_name,
    user_input
)

print("\n🧠 AI PRODUCTIVITY SUITE")
print("-----------------------")

print(result)

print("-----------------------")
print("✅ Done")
```
